```python
def potencia(base, expoente):
    """
    Calcula a pot√™ncia de um n√∫mero mostrando
    o conceito de repeti√ß√£o do fator.
    """
    
    resultado = 1  # elemento neutro da multiplica√ß√£o
    
    for _ in range(expoente):
        resultado *= base  # repete a multiplica√ß√£o da base
    
    return resultado


# Entrada de dados
base = int(input("Digite a base: "))
expoente = int(input("Digite o expoente: "))

# C√°lculo usando la√ßo (conceito matem√°tico)
resultado_repeticao = potencia(base, expoente)

# C√°lculo usando operador de pot√™ncia
resultado_operador = base ** expoente

# Sa√≠da
print(f"{base}^{expoente} (por repeti√ß√£o) = {resultado_repeticao}")
print(f"{base}^{expoente} (operador **) = {resultado_operador}")
```
