## Maior em Código

### Sobre a Matemática

Em matemática, **“maior”** é um conceito usado para comparar dois ou mais elementos e indicar qual deles possui valor mais elevado.

A tabela é uma forma organizada de apresentar dados — números, informações ou resultados — de modo que fiquem claros, comparáveis e fáceis de analisar. Ela é formada por linhas e colunas, onde cada célula representa um valor ou informação específica.

### Sobre o Código

Na linguagem de programação Python,

import – é a declaração que permite o acesso a módulos, pacotes ou funções de bibliotecas externas, tornando-os disponíveis no seu código para uso durante a execução.

pandas – é uma biblioteca utilizada para manipulação e análise de dados. Ela trabalha com DataFrame, que é uma estrutura de dados bidimensional (uma espécie de tabela com linhas e colunas.

as – é a expressão utilizada para criar um apelido (alias) para a biblioteca e ser usado quando for referenciada no código. pd é o alias comum para a biblioteca Pandas.

dicionário – é uma coleção mutável de pares chave-valor, onde cada chave é única. O dicionário é delimitado por chaves { }.

print – é a função utilizada para exibir mensagens na tela.

. – o ponto é um operador de acesso – ele serve para acessar algo que pertence a um objeto, como métodos, atributos ou submódulos.

max – é a função que localiza o maior elemento do conjunto.

'#' – é o caractere usado para comentar o código, não sendo executado pelo interpretador Python.

Vamos ao código…

```python
import pandas as pd

# Criando um DataFrame (tabela)
dados = {
    "Nome": ["Ana", "João", "Maria", "Pedro"],
    "Idade": [23, 21, 25, 20],
    "Nota": [8.5, 7.0, 9.0, 6.5]
}

df = pd.DataFrame(dados)

print(df)

# Maior nota
print("Maior nota:", df["Nota"].max())
```
