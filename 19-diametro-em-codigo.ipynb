## Diâmetro em Código

### Sobre a Matemática

Diâmetro é o segmento de reta que liga dois pontos de um circunferência e passa pelo seu centro. A relação fundamental é  d=2.r, onde r é o raio.

### Sobre o Código

Na linguagem de programação **Python**,

**def** - é a palavra-chave usada para definir uma função em Python, ou seja, um bloco de código reutilizável que executa uma tarefa específica.

**=** - é o operador de atribuição que associa (atribui) um valor a uma variável.

' * - é o operador de multiplicação que realiza o produto entre dois valores numéricos.

**return** - é a palavra-chave que retorna um valor da função para quem a chamou, encerrando a execução da função.

**float()** - é a função que converte um valor para número real (decimal), permitindo cálculos com casas decimais.

**input()** - é a função de entrada de dados (STDIN) que recebe informações digitadas pelo usuário em forma de texto.

**f"..."** (f-string) - é a forma de formatação de strings que permite inserir valores de variáveis diretamente no texto usando {}.

**'#'** - é usado para criar comentários ao código, que são notas para se entender melhor o que o código faz.

Vamos ao código...

```python
def calcular_diametro(raio):
    """
    Calcula o diâmetro de uma circunferência a partir do raio.
    
    Parâmetro:
    raio (float): valor do raio da circunferência
    
    Retorno:
    float: diâmetro da circunferência
    """
    diametro = 2 * raio
    return diametro


# Entrada de dados
raio = float(input("Digite o valor do raio da circunferência: "))

# Processamento
resultado = calcular_diametro(raio)

# Saída de dados
print(f"O diâmetro da circunferência é {resultado}")
```
