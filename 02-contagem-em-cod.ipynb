## Contagem de Valores em Código

### Sobre a Matemática

Contar valores significa determinar quantos elementos existem em um conjunto, sequência ou situação. A contagem é uma das ideias fundamentais da matemática — ela serve de base para operações, estatística, probabilidade e análise de dados.

### Sobre o Código

Na linguagem de programação Python,

list – é um tipo de sequência mutável. São coleções de itens que podem ser de qualquer tipo de dado. A lista é delimitada por colchetes [ ].

dicionário – é uma coleção mutável de pares chave-valor, onde cada chave é única. O dicionário é delimitado por chaves { }.

for – é o laço utilizado para iterar por intervalos definidos ou sequências de valores.

in – é o operador que pesquisa dentro do laço um determinado valor. Esse operador retorna True ou False.

print – a função é utilizada para exibir mensagens na tela.

'#' – é o caractere usado para comentar o código, não sendo executado pelo interpretador Python.

Vamos ao código…

```python
# Quantos alunos escolheram cada esporte
esportes = ["futebol", "basquete", "futebol", "natação", "futebol", "natação"]

contagem = {}
for esporte in esportes:
    contagem[esporte] = contagem.get(esporte, 0) + 1

print("Participação por esporte:", contagem)
```
