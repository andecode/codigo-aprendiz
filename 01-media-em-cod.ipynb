## Média em Código

### Sobre a Matemática

A média de um conjunto de dados é encontrada somando-se todos os números do conjunto de dados e então dividindo o resultado pelo número de valores do conjunto. Assim, a média aritmética de dois números a e b é a soma a + b dividida por 2; a média aritmética de três números a, b e c é a soma de a + b + c dividida por 3, etc.

### Sobre o Código

Na linguagem de programação Python,

list – é um tipo de sequência mutável. A lista é delimitada por colchetes [ ]. Nó código abaixo, idades é uma variável do tipo list.

sum – é a função utilizada para somar os elementos.

len – é a função utilizada para obter o número de elementos.

/ – é o operador de divisão.

print – é a função utilizada para exibir mensagens na tela.

'#' – é o caractere usado para comentar o código, não sendo executado pelo interpretador Python.

Vamos ao código…

```python
# Lista com idades de alunos
idades = [15, 16, 15, 17, 16, 18, 15]

# Média
media = sum(idades) / len(idades)

print("Idades:", idades)
print("Média das idades:", media)
```
