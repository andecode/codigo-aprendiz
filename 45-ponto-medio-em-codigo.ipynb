```python
# Ponto m√©dio usando biblioteca sympy
from sympy import Point, Segment

# Definindo os pontos extremos do segmento
A = Point(2, 4)
B = Point(6, 8)

# Criando o segmento de reta
segmento = Segment(A, B)

# Calculando o ponto m√©dio
ponto_medio = segmento.midpoint

print(f"O ponto m√©dio do segmento √©: {ponto_medio}")
```

|---|---|
|from sympy import Point, Segment|Importa da biblioteca *SymPy* as classes *Point* (ponto) e *Segment (segmento de reta). |

|B = Point(6, 8)| Cria o ponto **B** com coordenadas (6, 8).|
|segmento = Segment(A, B)| Cria um segmento de reta ligando o ponto A ao ponto B.|
|ponto_medio = segmento.midpoint| Calcula automaticamente o ponto m√©dio do segmento usando a propriedade *midpoint*.|
|print(f"O ponto m√©dio do segmento √©: {ponto_medio}")| Exibe o ponto m√©dio na tela usando *f-string*.|

```python
# Ponto m√©dio sem biblioteca (f√≥rmula matem√°tica)
# Coordenadas dos pontos
x1, y1 = 2, 4
x2, y2 = 6, 8

# C√°lculo do ponto m√©dio
xm = (x1 + x2) / 2
ym = (y1 + y2) / 2

print(f"O ponto m√©dio do segmento √©: ({xm}, {ym})")
```
