```python
def calcular_fracao():
    """
    Calcula e interpreta uma fra√ß√£o a partir do numerador e do denominador.

    Numerador  ‚Üí quantidade de partes consideradas
    Denominador ‚Üí quantidade total de partes iguais do todo
    """

    # Entrada de dados
    numerador = int(input("Digite o numerador (partes consideradas): "))
    denominador = int(input("Digite o denominador (partes do todo): "))

    # Verifica√ß√£o matem√°tica
    if denominador == 0:
        print("‚ùå O denominador n√£o pode ser zero, pois n√£o existe divis√£o por zero.")
    else:
        # C√°lculo da fra√ß√£o
        valor_fracao = numerador / denominador

        # Sa√≠da de dados
        print(f"\nüìê Fra√ß√£o formada: {numerador}/{denominador}")
        print(f"üîπ O todo foi dividido em {denominador} partes iguais.")
        print(f"üîπ Foram consideradas {numerador} dessas partes.")
        print(f"üîπ Valor num√©rico da fra√ß√£o: {valor_fracao}")

# Chamada da fun√ß√£o
calcular_fracao()
```

```python
from fractions import Fraction
import math

def trabalhar_fracao():
    """
    Trabalha com o conceito matem√°tico de fra√ß√£o utilizando bibliotecas Python.

    Numerador   ‚Üí partes consideradas
    Denominador ‚Üí partes iguais do todo
    """

    # Entrada de dados
    numerador = int(input("Digite o numerador (partes consideradas): "))
    denominador = int(input("Digite o denominador (partes do todo): "))

    # Verifica√ß√£o matem√°tica
    if denominador == 0:
        print("‚ùå O denominador n√£o pode ser zero.")
    else:
        # Cria√ß√£o da fra√ß√£o usando biblioteca
        fracao = Fraction(numerador, denominador)

        # Convers√µes e c√°lculos
        valor_decimal = float(fracao)
        valor_absoluto = math.fabs(valor_decimal)

        # Sa√≠da de dados
        print(f"\nüìê Fra√ß√£o informada: {numerador}/{denominador}")
        print(f"üîπ Fra√ß√£o simplificada (biblioteca): {fracao}")
        print(f"üîπ Valor decimal da fra√ß√£o: {valor_decimal}")
        print(f"üîπ Valor absoluto da fra√ß√£o: {valor_absoluto}")

# Chamada da fun√ß√£o
trabalhar_fracao()
```
